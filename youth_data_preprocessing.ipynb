```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/2024_총괄_20250722_25735.csv', encoding='euc-kr')
df
```

```python
df.info()
```

```python
df.columns 
```

```python
# 장애 '해당없음(0)' 만 추출
df = df[df['장애여부'] == 0]
df
```

```python
# 대한민국 국적만 추출
df = df[(df['국적'] == 1) | (df['(복수 응답) 현재 국적_1'] == 1)]
df
```

```python
# 건강, 신체, 정신 장애로 사회활동에 제한을 받지 않는 사람만 추출
df = df[df['건강 문제 혹은 장애로 인한 일상생활 및 사회활동을 제한을 받았던 경험'] == 2]
df
```

```python
# 1주일 동안 1시간 이상 일한 사람(취업자 기준)만 추출
# df = df[df['지난 주 수입을 목적으로 1시간 이상 일한 경험'] == 1]
# df
```

```python
# 직장에서 '근로자'로 일한 사람만 추출
df = df[(df['지난 주 종사상 지위'] == 1) | (df['지난 주 종사상 지위'] == 2) | (df['지난 주 종사상 지위'] == 3)]
df
```

```python
# 필요 없는 열 제거
df = df.drop(['장애여부', '국적', '(복수 응답) 현재 국적_1', '(복수 응답) 현재 국적_2', 
              '귀화 여부', '건강 문제 혹은 장애로 인한 일상생활 및 사회활동을 제한을 받았던 경험',
              '취업 준비 도움 주체 - 학교(상담 교사^ 교사^ 지도 교수 등)', '취업 준비 도움 주체 - 사설 기관(사설 학원^ 인터넷 취업 사이트 등)',
              '취업 준비 도움 주체 - 부모님이나 친인척', '취업 준비 도움 주체 - 친구^ 선후배', '취업 준비 도움 주체 - 온라인 커뮤니티/SNS. 유튜브',
              '취업 준비 도움 정도 - 학교(상담 교사^ 교사^ 지도 교수 등)', '취업 준비 도움 정도 - 사설 기관(사설 학원^ 인터넷 취업 사이트 등)',
              '취업 준비 도움 정도 - 부모님이나 친인척', '취업 준비 도움 정도 - 온라인 커뮤니티/SNS. 유튜브',
              '취업 준비 도움 정도 - 친구^ 선후배', 
              '최근 1년간 취업 및 창업 경험 여부', '지난 4주 이내 구직 경험',
              ], axis=1)
df
```

```python
# 컬럼명 정의
df.columns = ['성별', '연령별', '지역별', '학력', '운동', '취업 준비 도움 주체 - 공공 기관', 
               '취업 준비 도움 정도 - 공공 기관', 
               '취업자여부', '고용형태', '고용기간', '사업체규모', '재직기간(범위)', 
              '가구 연간소득 - 총 소득(만원)', '수혜여부(만원)', '부채(만원)', '재산(만원)']
df
```

```python
# 결측치 확인
df.isna().sum()
```

```python
# 취업 준비 도움 주체, 도움 정도 결측치 -> 없다(2)로 채우기
fillna_cols = [
    '취업 준비 도움 주체 - 공공 기관', '취업 준비 도움 정도 - 공공 기관'
]
df[fillna_cols] = df[fillna_cols].fillna(2)

# float -> int로 바꾸기
df[fillna_cols] = df[fillna_cols].astype(int)
df
```

```python
# float 타입 열 -> int 타입으로 변환
float_to_int_cols = [
    '고용형태', 
    '고용기간', 
    '사업체규모', 
    '재직기간(범위)'
]

df[float_to_int_cols] = df[float_to_int_cols].astype(int)
df
```

```python
# 결측치 없음 확인
df.isna().sum()
```
